После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
a = data.groupby('occupation')['age'].count()
b = data.groupby('occupation')['age'].mean()
c = data.groupby('occupation')['hours-per-week'].min()


d = pd.merge(a, b, on='occupation')
e = pd.merge(d, c, on='occupation')


e

def filter_func(row):
    if row['age_y'] < 40 and row['hours-per-week'] > 5:
        return True
    else:
        return False
    
e.loc[e.apply(filter_func, axis=1) == True]

```
